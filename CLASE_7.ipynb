```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datos/ml_project1_data.csv')
```

```python
df.columns
```

```python
df = pd.get_dummies(df, columns=['Education','Marital_Status'])
```

```python
df
```

```python
df.describe()
```

```python
df.hist(figsize=(25, 10), bins=50, edgecolor='black')
plt.subplots_adjust(hspace=0.6, wspace=0.5)
plt.show()
```

```python
plt.figure(figsize=(32, 12))
sns.heatmap(df.drop(columns=['Dt_Customer']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df.columns
```

```python
plt.figure(figsize=(24, 12))
plt.subplot(2,3,1)
sns.boxplot(x='Response', y='Income', data=df)
#plt.subplots_adjust(top=0.9)
plt.title('Income Distribution by Response')  
plt.subplot(2,3,2)
sns.boxplot(x='Response', y='Marital_Status', data=df)
plt.title('Marital Distribution by Response')
plt.subplot(2,3,3)
sns.boxplot(x='Response', y='Education', data=df)
plt.title('Education Distribution by Response')
plt.tight_layout()
plt.subplots_adjust(top=0.9)   
plt.show()
```
