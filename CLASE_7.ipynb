```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('datos/ml_project1_data.csv')
```

```python
df.columns
```

```python
df = pd.get_dummies(df, columns=['Education','Marital_Status'])
```

```python
df
```

```python
df.describe()
```

```python
df.hist(figsize=(25, 10), bins=50, edgecolor='black')
plt.subplots_adjust(hspace=0.6, wspace=0.5)
plt.show()
```

```python
plt.figure(figsize=(32, 12))
sns.heatmap(df.drop(columns=['Dt_Customer']).corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()
```
